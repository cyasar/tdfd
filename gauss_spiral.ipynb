```python
import matplotlib.pyplot as plt
import numpy as np

# Asal sayı kontrolü
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# Gauss asal sayı kontrolü
def is_gaussian_prime(a, b):
    if a != 0 and b != 0:
        return is_prime(a**2 + b**2)
    elif a == 0:
        return is_prime(abs(b)) and abs(b) % 4 == 3
    elif b == 0:
        return is_prime(abs(a)) and abs(a) % 4 == 3

# Grafik oluşturma
x_vals = []
y_vals = []

limit = 100  # Değerleri kontrol etmek için kullanılacak limit

for a in range(-limit, limit + 1):
    for b in range(-limit, limit + 1):
        if is_gaussian_prime(a, b):
            x_vals.append(a)
            y_vals.append(b)

plt.figure(figsize=(10, 10))
plt.scatter(x_vals, y_vals, s=5, c='blue')
plt.title("Gaussian Prime Spirals")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def is_gaussian_prime(a, b):
    if a != 0 and b != 0:
        return is_prime(a**2 + b**2)
    elif a == 0:
        return is_prime(abs(b)) and abs(b) % 4 == 3
    elif b == 0:
        return is_prime(abs(a)) and abs(a) % 4 == 3

center_a, center_b = -12, -7
limit = 5  # Spiral için yarıçap
points = []

for a in range(center_a - limit, center_a + limit + 1):
    for b in range(center_b - limit, center_b + limit + 1):
        if is_gaussian_prime(a, b):
            points.append((a, b))

# Sort the points in spiral fashion
points.sort(key=lambda x: (x[0] - center_a)**2 + (x[1] - center_b)**2)

x_vals = [point[0] for point in points]
y_vals = [point[1] for point in points]

plt.figure(figsize=(10, 10))
plt.plot(x_vals, y_vals, marker="o", linestyle="-", c='blue')
plt.scatter([center_a], [center_b], s=100, c='red')  # Merkezi kırmızı ile işaretle
plt.title("Gaussian Prime Spiral around -12 - 7i")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid(True)
plt.show()
```
